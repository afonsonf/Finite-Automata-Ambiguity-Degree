```python
from FAdo import *
from FAdo.reex import *
from FAdo.fio import *
```

```python
r = str2regexp("ab")
m = r.toNFA()
m.display()
```

```python
r1 = str2regexp("ab+aa*b")
m1 = r1.toNFA()
m1.display()
```

```python
m.renameStates()
m1.renameStates()
prod1 = m.product(m1)
for x in [(m.States[a], m1.States[b]) for a in m.Final for b in m1.Final]:
            if x in prod1.States:
                prod1.addFinal(prod1.stateIndex(x))
prod1.trim().display(strict = True)
```
