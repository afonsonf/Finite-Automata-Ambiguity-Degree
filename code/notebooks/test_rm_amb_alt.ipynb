```python
from FAdo.fa import *
from FAdo.reex import *
from FAdo.fio import *
from FAdo.cfg import *
```

```python
#aa*ba + aba*

m1 = NFA()
for i in range(0,7):
    m1.addState()
m1.addInitial(0)
m1.addFinal(3)
m1.addFinal(6)

m1.addTransition(0,'b',1)
m1.addTransition(1,'a',1)
m1.addTransition(1,'b',2)
m1.addTransition(2,'a',3)

m1.addTransition(0,'b',4)
m1.addTransition(4,'a',5)
m1.addTransition(5,'b',6)
m1.addTransition(6,'a',6)


m1.renameStates()
m1.display()
prod1 = m1.product(m1)
for x in [(m1.States[a], m1.States[b]) for a in m1.Final for b in m1.Final]:
            if x in prod1.States:
                prod1.addFinal(prod1.stateIndex(x))
prod1.trim().display()
```

```python
#baba caminho ambiguo x inicial

m2 = NFA()
for i in range(0,5):
    m2.addState()
m2.addInitial(0)
m2.addFinal(4)

m2.addTransition(0,'b',1)
m2.addTransition(1,'a',2)
m2.addTransition(2,'b',3)
m2.addTransition(3,'a',4)


m2.renameStates()
m2.display()
m1.display()
prod2 = m1.product(m2)
for x in [(m1.States[a], m2.States[b]) for a in m1.Final for b in m2.Final]:
            if x in prod2.States:
                prod2.addFinal(prod2.stateIndex(x))
prod2.trim().display()
```
