```python
s = str2regexp(gen.generate(),parser=ParserRPN)
nfa = s.toNFA()

sz = len(nfa.States)
nfa.addState()

for s in nfa.Final: nfa.addTransition(s,'#',sz)
nfa.Final = {sz}

print ambiguityTest.isUFA(nfa)
print ambiguityTest.isFNFA(nfa)
print ambiguityTest.isPNFA(nfa)

nfa.renameStates()
nfa.display()
prod = nfa.product(nfa)
for x in [(nfa.States[a], nfa.States[b]) for a in nfa.Final for b in nfa.Final]:
    if x in prod.States:
        prod.addFinal(prod.stateIndex(x))
prod.trim().display()
```
