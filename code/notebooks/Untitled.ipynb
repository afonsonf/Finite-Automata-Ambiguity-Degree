```python
r = str2regexp("a*bb + aab* + aa*bb* + a*b*")
m = r.toNFA().trim()
# m.toDFA().display()
m.renameStates()
m.display()
prod = m.product(m)
for x in [(m.States[a], m.States[b]) for a in m.Final for b in m.Final]:
            if x in prod.States:
                prod.addFinal(prod.stateIndex(x))
prod.trim().display()
```
