```python
from FAdo.reex import *
from FAdo.fio import *
import ambiguityTest
```

```python
m5 = NFA()
for i in range(0,6):
    m5.addState()
m5.addInitial(0)
m5.addFinal(5)

m5.addTransition(0,'u',4)
m5.addTransition(4,'w',5)

m5.addTransition(4,'v',1)
m5.addTransition(1,'x',2)
m5.addTransition(1,'x',3)
m5.addTransition(2,'x',4)
m5.addTransition(3,'x',4)

m5.display()
print ambiguityTest.isPNFA(m5)
print ambiguityTest.isFNFA(m5)
print ambiguityTest.isUFA(m5)

# x5 = m5.conjunction(m5)
# x5.display(strict = True)
```

```python
m1 = NFA()
for i in range(0,6):
    m1.addState()
m1.addInitial(0)
m1.addFinal(3)

m1.addTransition(0,'c',1)
m1.addTransition(0,'c',4)

m1.addTransition(1,'a',1)
m1.addTransition(1,'b',2)
m1.addTransition(2,'b',3)

m1.addTransition(4,'a',5)
m1.addTransition(5,'a',3)
m1.addTransition(3,'b',3)

m1.renameStates()
m1.display()
print ambiguityTest.isPNFA(m1)
print ambiguityTest.isFNFA(m1)
print ambiguityTest.isUFA(m1)

x1 = m1.conjunction(m1)
x1.trim()
x1.display(strict = True)
```
